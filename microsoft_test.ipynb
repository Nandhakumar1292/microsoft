# Data Cleaning

```python
df_test.duplicated().sum()
```

```python
df_test = df_test.drop_duplicates()
```

```python
df_test.duplicated().sum()
```

```python
df_test.isnull().sum()
```

```python
df_test['Timestamp'] = pd.to_datetime(df_test['Timestamp'])
```

```python
df_test['Timestamp'].dtype
```

```python
df_test['Day'] = df_test['Timestamp'].dt.day
df_test['Month'] = df_test['Timestamp'].dt.month
df_test['Year'] = df_test['Timestamp'].dt.year
df_test['Hour'] = df_test['Timestamp'].dt.hour
df_test['Time'] = df_test['Timestamp'].dt.time
```

```python
df_test
```

```python
df_test.columns
```

```python
# Removing columns based on the trained cleaned csv
df_test = df_test.drop(["Sha256","AccountSid","AccountUpn","AccountObjectId","RegistryValueName","ApplicationId","FolderPath",
                "CountryCode","State","Month", "OSFamily", "DeviceId" ,"Timestamp",
               "MitreTechniques","ActionGrouped","ActionGranular","EmailClusterId","ThreatFamily","ResourceType","Roles",
                "AntispamDirection","SuspicionLevel","LastVerdict","Usage"], axis =1)
```

```python
df_test.columns
```

```python
df_test.info()
```

```python
categorical_cols = df_test.select_dtypes(include=['object']).columns
categorical_cols
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for column in categorical_cols:
    df_test[column] = label_encoder.fit_transform(df_test[column])
```
